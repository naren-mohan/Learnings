```python
country_grp = df.groupby('Country')
#country_grp['LanguageWorkedWith'].value_counts()
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
pyth_users = country_grp['LanguageWorkedWith'].agg(['count', lambda x: x.str.contains('Python').sum()])
pyth_users = pyth_users.rename(columns = {'count':'Total Users', '<lambda_0>':'Python Users'})
pyth_users['Pyth_Percentage'] = pyth_users['Python Users'] * 100 / pyth_users['Total Users']
min_users = pyth_users['Total Users'] > 100
pyth_users[min_users].sort_values(by='Pyth_Percentage', ascending=False) 
```
